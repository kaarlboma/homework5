```python
import numpy as np
import matplotlib.pyplot as plt
from Exp3 import Exp3
from UCB import UCB
```

```python
# Part a
delta = 0.05
mu1 = 0.5
mu2 = mu1 + delta
Mu = [mu1, mu2]
num_simulations = 20
K = len(Mu)
delta = 0.01
ns = np.arange(100, 10001, 100)
etas = np.sqrt((2 * np.log(K)) / (ns * K))
exp3_regrets = []
ucb_regrets = []
```

```python
for i in range(len(ns)):
    eta = etas[i]
    n = ns[i]
    exp3_regret = Exp3(Mu, eta, n, num_simulations)
    ucb_regret = UCB(delta, Mu, n, num_simulations)
    exp3_regrets.append(exp3_regret)
    ucb_regrets.append(ucb_regret)
```

```python
colors = {
    "UCB": "#1b9e77",
    "Exp3": "#d95f02",
}
plt.figure(figsize=(8,6), dpi=150)

plt.plot(ns, exp3_regrets, label="Exp3", color=colors["Exp3"], linewidth=2)
plt.plot(ns, ucb_regrets, label="UCB", color=colors["UCB"], linewidth=2)

plt.xlabel("Horizon n", fontsize=14)
plt.ylabel("Expected Regret", fontsize=14)
plt.title("Expected Regret vs Horizon n", fontsize=16)
plt.legend(fontsize=12, frameon=True)
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# Part b
n = 10 ** 5
etas = np.linspace(0.001, 0.15, 50)
exp3_regrets = []
num_simulations = 50

for i in range(len(etas)):
    eta = etas[i]
    exp3_regret = Exp3(Mu, eta, n, num_simulations)
    exp3_regrets.append(exp3_regret)
```

```python
colors = {
    "UCB": "#1b9e77",
    "Exp3": "#d95f02",
}
plt.figure(figsize=(8,6), dpi=150)

plt.plot(etas, exp3_regrets, label="Exp3", color=colors["Exp3"], linewidth=2)

plt.xlabel("Learning Rate", fontsize=14)
plt.ylabel("Expected Regret", fontsize=14)
plt.title("Expected Regret vs Learning Rate", fontsize=16)
plt.legend(fontsize=12, frameon=True)
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
# Part bc
n = 10 ** 5
etas = np.linspace(0.001, 0.15, 50)
exp3_regrets = []
num_simulations = 50
Mu = [mu1, mu2 + 0.4]

for i in range(len(etas)):
    eta = etas[i]
    exp3_regret = Exp3(Mu, eta, n, num_simulations)
    exp3_regrets.append(exp3_regret)
```

```python
colors = {
    "UCB": "#1b9e77",
    "Exp3": "#d95f02",
}
plt.figure(figsize=(8,6), dpi=150)

plt.plot(etas, exp3_regrets, label="Exp3", color=colors["Exp3"], linewidth=2)

plt.xlabel("Learning Rate", fontsize=14)
plt.ylabel("Expected Regret", fontsize=14)
plt.title("Expected Regret vs Learning Rate", fontsize=16)
plt.legend(fontsize=12, frameon=True)
plt.grid(alpha=0.2)
plt.tight_layout()
plt.show()
```

UCB and Exp3 both achieve sublinear regret but in this setting with a $\Delta = 0.05$, Exp3 outperforms UCB achieving lower expected regret as the horizon $n$ increases. One reason could be because of Exp3's more aggressive exploration in earlier rounds, the importance weighting helps it to outperform UCB. UCB could also have too wide of confidence bounds which could cause it to over-explore.
